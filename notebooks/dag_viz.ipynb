### Model 2

```python
# Instantiate the PGM.
pgm = daft.PGM([6, 5], origin=[0, 0])

#Hierarchical parameters
pgm.add_node(daft.Node('mu_skill', r"$\mu_{skill}$", 1, 4.5, fixed=True))
pgm.add_node(daft.Node('tau_skill', r"$\tau_{skill}$", 1.5, 4.5, fixed=True))

#Latent variables
pgm.add_node(daft.Node("skill", r"skill$_i$", 1.25, 3.5, aspect=1.9))
pgm.add_node(daft.Node("injury", r"injury", 3, 3.5, aspect=1.9))
pgm.add_node(daft.Node("corner", r"corner", 4, 3.5, aspect=1.9))

#Data
pgm.add_node(daft.Node("y_arc_attempts", r"attempts$_{arc, i, j}$", 1, 1.5, observed=True, scale=1.5, aspect=1.9))
pgm.add_node(daft.Node("y_arc_made", r"y$_{arc, i, j}$", 2, .5, observed=True, aspect=1.9))
pgm.add_node(daft.Node("injured", r"injured$_{i,j}$", 3, 2.5, observed=True, aspect=1.9))

pgm.add_node(daft.Node("y_corner_attempts", r"attempts$_{corner, i, j}$", 5, 1.5, observed=True, scale=1.7, aspect=1.9))
pgm.add_node(daft.Node("y_corner_made", r"y$_{corner, i, j}$", 4, .5, observed=True, aspect=1.9))


#Add in the edges.
pgm.add_edge("mu_skill", "skill")
pgm.add_edge("tau_skill", "skill")
pgm.add_edge("skill", "y_arc_made")
pgm.add_edge("skill", "y_corner_made")
pgm.add_edge("y_arc_attempts", "y_arc_made")
pgm.add_edge("y_corner_attempts", "y_corner_made")
pgm.add_edge("injured", "y_arc_made")
pgm.add_edge("injured", "y_corner_made")
pgm.add_edge("injury", "y_arc_made")
pgm.add_edge("injury", "y_corner_made")

pgm.add_edge("corner", "y_corner_made")

#Render and save.
pgm.render()
#pgm.figure.savefig(GRAPHDIR + "model2.pdf")
pgm.figure.savefig(GRAPHDIR + "model2.png", dpi=150)
```
