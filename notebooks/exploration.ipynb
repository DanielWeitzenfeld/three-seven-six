# Top 3 point shooters last season

```python
q = session.query(Season).options(joinedload(Season.totals),
                                  joinedload(Season.player),                                  
                                  ).filter(Season.season_start == 2013).all()

def make_row(s):
    #print s.player.player
    d = {'player': s.player.player,
         }
    for col in ['tpa', 'tpm', 'fgpc_3p', 'mid_season_team_switch', 'team', 'teams', 
                'positions', 'arc_3pfga', 'corner_3pfga', 'fgpc_3parc', 'fpgc_3pcorner']:
        d['%s_2013' % col] = getattr(s, col)
    if s.prior_season:
        for col in ['tpa', 'tpm', 'fgpc_3p', 'mid_season_team_switch', 'team', 'teams']:
            d['%s_2012' % col] = getattr(s.prior_season, col)  
    return d

df = pd.DataFrame([make_row(s) for s in q])
df['percentage_point_change'] = df.fgpc_3p_2013 - df.fgpc_3p_2012
df['percentage_change_in_attempts'] = (df.tpa_2013 - df.tpa_2012) / df.tpa_2012
```

```python
df.head()
```

## Biggest changes: 2012-13 to 2013-14

```python
df = df.sort_index(by='fgpc_3p_2013', ascending=False)
sub_frame = df.ix[(df.tpa_2013 > 80) & 
                  (df.tpa_2012 > 80), ['player', 'tpa_2012', 'teams_2012', 'fgpc_3p_2012', 
                                       'tpa_2013', 'fgpc_3p_2013',  'teams_2013', 
                                       'percentage_point_change', 'percentage_change_in_attempts']]
sub_frame.to_csv(three_seven_six.ROOT + '/data/at_least_80_attempts.csv', index=False)
sub_frame.head()
```

## Is everyone better at the corner 3 than the ex-corner 3?

```python
MIN_ATTEMPTS = 60
sub_frame = df.ix[(df.corner_3pfga_2013 > MIN_ATTEMPTS) & 
                  (df.arc_3pfga_2013 > MIN_ATTEMPTS), ['player', 'positions_2013', 'teams_2013', 
                                                       'fgpc_3parc_2013', 'corner_3pfga_2013', 'fpgc_3pcorner_2013']]
sub_frame['worse_at_corner'] = (sub_frame.fgpc_3parc_2013 > sub_frame.fpgc_3pcorner_2013)
sub_frame.head()
```
