```python
import seaborn as sns
import pandas as pd
import three_seven_six
from three_seven_six.dbs import mysql as mysql_db
from three_seven_six.models import *

from sqlalchemy.orm import joinedload, subqueryload 

session = mysql_db.Session()
```

# Top 3 point shooters last season

```python
q = session.query(Season).options(joinedload(Season.shooting),
                                  joinedload(Season.totals),
                                  joinedload(Season.player),
                                  
                                  ).filter(Season.season_start == 2013, 
                                                                      Season.mid_season_team_switch == False).all()

def make_row(r):
    d = {'player': r.player.player,
         }
    if r.shooting:
        d.update(r.shooting[0].to_dict())
        d.update(r.totals[0].to_dict())
    return d

df = pd.DataFrame([make_row(r) for r in q])
```
